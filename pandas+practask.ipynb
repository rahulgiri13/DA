```python
import pandas as pd
```

```python
file__path=  'D:/arc classes/Paython/pandas/ECOM.xlsx'
```

```python
ecom_data = pd.read_excel('D:/arc classes/Paython/pandas/ECOM.xlsx',sheet_name=['customer', 'Sales', 'Product'],header=0)
```

```python
ecom_data
```

```python
top50_customer = ecom_data['customer'].head(50).to_dict(orient='records')

top50_sales = ecom_data['Sales'].head(50).to_dict(orient='records')

top50_product = ecom_data['Product'].head(50).to_dict(orient='records')
```

```python
print(f'top50_customer = {top50_customer}')
print()
print(f'top50_sales = {top50_sales}')
print()
print(f'top50_product = {top50_product}')
```

```python
customer_data = pd.DataFrame(top50_customer)
print()
sales_data = pd.DataFrame(top50_sales)
print()
product_data = pd.DataFrame(top50_product)
```

```python
print(f'customer_data = {customer_data}')
```

```python
print(f'sales_data = {sales_data}')
```

```python
print(f'product_data = {product_data}')
```

```python
customer_data.to_csv('customer_top50.csv', index=0)
sales_data.to_csv('sales_top50.csv', index=0)
product_data.to_csv('product_top50.csv', index=0)
```

```python
with pd.ExcelWriter('ecom_top50_data.xlsx') as writer:
    customer_data.to_excel(writer, sheet_name='customer_top50', index=0)
    sales_data.to_excel(writer, sheet_name='Sales_top50', index=0)
    product_data.to_excel(writer, sheet_name='product_top50', index=0)
```

```python
customer_aggregation = customer_data.describe()

print(f'Customer Aggregations = {customer_aggregation}')
```
